# Performing a T-Test for 2017 Income using CA and CT as our sample populations

```python
CT_2017 = t_test_2017_df[t_test_2017_df['State'] == 'Connecticut']
CT_2017.Income.mean()
```

```python
CA_2017 = t_test_2017_df[t_test_2017_df['State'] == 'California']
CA_2017.Income.mean()
```

```python
stats.ttest_ind(CT_2017.Income, CA_2017.Income, equal_var=False)
```

#### P-Value (pvalue=0.00458 < 0.05) therefore, we reject the null hypothesis. There is a statistical difference between Conneticut and California's populations average incomes. This makes sense since California has a far larger population and wider scope of workforce types (that impacts the state's income average).
